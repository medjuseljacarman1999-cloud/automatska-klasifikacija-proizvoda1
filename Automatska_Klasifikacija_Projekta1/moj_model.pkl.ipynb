```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
import joblib

# 1. Učitavanje
df = pd.read_csv('IMLP6_TASK_03-products (2).csv')

# 2. Čišćenje naziva kolona (OVO REŠAVA TVOJ PROBLEM SA RAZMACIMA)
df.columns = df.columns.str.strip()

# 3. Priprema podataka
df = df.dropna(subset=['Product Title', 'Category Label'])
X = df['Product Title']
y = df['Category Label']

# 4. Kreiranje Pipeline-a
model = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english')),
    ('clf', MultinomialNB())
])

# 5. Treniranje
model.fit(X, y)

# 6. Čuvanje modela
joblib.dump(model, 'moj_model.pkl')
print("Model je uspešno istreniran i sačuvan kao 'moj_model.pkl'!")
```

```python
import joblib

# Učitaj sačuvani model
model = joblib.load('moj_model.pkl')

print("Sistem za klasifikaciju proizvoda spreman!")
print("Upišite 'kraj' za izlaz.")

# Interaktivno testiranje
while True:
    naslov = input("\nUnesite naziv proizvoda: ")
    if naslov.lower() == 'kraj':
        break

    # Predikcija
    predikcija = model.predict([naslov])
    print(f"Predviđena kategorija: {predikcija[0]}")
```
